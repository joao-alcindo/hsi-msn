```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import glob
```

```python
normalized_npy = './../../hyperview1/normalized_w_o_patches_npy/train/'
normalized_test_npy = './../../hyperview1/normalized_w_o_patches_npy/test/'

normalized_train_files = glob.glob(normalized_npy + '/*.npy')
normalized_test_files = glob.glob(normalized_test_npy + '/*.npy')
```

```python
# sort files to ensure matching order
normalized_train_files.sort()
normalized_test_files.sort()
```

```python
# resize usando nearest neighbor, use torch for that
import torch
def resize_nearest_neighbor(image, new_size):
    image_tensor = torch.from_numpy(image).unsqueeze(0).unsqueeze(0)  # add batch and channel dimensions
    resized_tensor = torch.nn.functional.interpolate(image_tensor, size=new_size, mode='nearest')
    return resized_tensor.squeeze().numpy()  # remove batch and channel dimensions

train_data = [resize_nearest_neighbor(np.load(f), (150,64, 64)) for f in normalized_train_files]
test_data = [resize_nearest_neighbor(np.load(f), (150,64, 64)) for f in normalized_test_files]
```

```python

six_random_channels = [10, 23, 45, 78, 101, 123]  # example channel indices to remove

train_data_144 = [np.delete(f, six_random_channels, axis=0) for f in train_data]
test_data_144 = [np.delete(f, six_random_channels, axis=0) for f in test_data]
```

```python
# save in resized_144 folder
import os
resized_144_train_npy = './../../hyperview1/resized_144/train/'
resized_144_test_npy = './../../hyperview1/resized_144/test/'

os.makedirs(resized_144_train_npy, exist_ok=True)
os.makedirs(resized_144_test_npy, exist_ok=True)


for i, f in enumerate(normalized_train_files):
    base_name = os.path.basename(f)
    np.save(os.path.join(resized_144_train_npy, base_name), train_data_144[i])

for i, f in enumerate(normalized_test_files):
    base_name = os.path.basename(f)
    np.save(os.path.join(resized_144_test_npy, base_name), test_data_144[i])
```

```python
import os
# save resized images 

resize_folder = './../../hyperview1/resized/'
if not os.path.exists(resize_folder + 'train/'):
    os.makedirs(resize_folder + 'train/')
if not os.path.exists(resize_folder + 'test/'):
    os.makedirs(resize_folder + 'test/')

for i, img in enumerate(train_data):
    np.save(resize_folder + 'train/resized_img_' + f'{i:05d}' + '.npy', img)

for i, img in enumerate(test_data):
    np.save(resize_folder + 'test/resized_img_' + f'{i:05d}' + '.npy', img)
```
