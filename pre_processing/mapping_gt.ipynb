```python
import pandas as pd
import glob
import os
```

```python
train_dir = './../hyperview1/pca/train/'
train_files = glob.glob(train_dir + '*.npy')

train_files.sort()
```

```python
# get the base names of the files
base_names = [int(os.path.basename(f).split('_')[0]) for f in train_files]

df = pd.DataFrame({'file': train_files, 'base_name': base_names})
```
